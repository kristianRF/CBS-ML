5) Show the data types of each column in df and convert the object columns to categorical variables

```python
df.dtypes
```

```python
object_list = list(df.select_dtypes(include="object").columns)
```

```python
df[object_list] = df.select_dtypes(include="object").astype("category")
df.dtypes
```

6) Summarise the numerical variables in df (e.g., number of observations, mean, standard deviations, min value, max value, etc.) and check if the numerical variables are linearly correlated

```python
df.describe()
```

```python
%matplotlib inline
x = df['Usage_Length'].to_numpy()
y = df['Monthly_Fee'].to_numpy()

m, b = np.polyfit(x, y, 1)

plt.plot(x, m * x + b, color='r')

plt.scatter(x, y, label=f'Correlation = {np.round(np.corrcoef(x,y)[0,1], 3)}')

plt.title('Scatterplot and correlations')
plt.legend()
plt.show()
```
